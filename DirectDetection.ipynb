```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as pat
from matplotlib.collections import PatchCollection
```

#Spin Dependent

```python
SD = pd.read_table("SD.csv",sep=",",header=None,names=range(100)).T
SD.columns = [SD[i][0] for i in SD]
SD = SD.ix[1:]
```

```python
SD.head()
```

```python
### Create figure and basic styling
fig = plt.figure(dpi=1200)
ax = fig.add_subplot(111,xlabel='WIMP mass [GeV]',ylabel='WIMP-nucleon cross-section [cm$^2$]')

for item in [ax.xaxis.label, ax.yaxis.label]:
    item.set_fontsize(16)

for item in (ax.get_xticklabels() + ax.get_yticklabels()):
    item.set_fontsize(14)

plt.xlim([5,10**4])
plt.ylim([10**-40,10**-33])

### Exclusion lines
cdms       = plt.loglog(10**SD['CDMS_MASS'],10**SD['CDMS_XSEC'],
                        label='CDMS',linewidth='2',color="teal")
xenon10    = plt.loglog(10**SD['XENON10_MASS'],10**SD['XENON10_XSEC'],
                        label='XENON10',linewidth='2',linestyle="--",color="tomato")
xenon100   = plt.loglog(10**SD['XENON100_MASS'],10**SD['XENON100_XSEC'],
                        label='XENON100',linewidth='2',color='tomato')
picasso    = plt.loglog(10**SD['PICASSO_MASS'],10**SD['PICASSO_XSEC'],
                        label='PICASSO',linewidth='2',color="olive")
simple     = plt.loglog(10**SD['SIMPLE_MASS'],10**SD['SIMPLE_XSEC'],
                        label='SIMPLE',linewidth='2',color="steelblue")
coupp      = plt.loglog(10**SD['COUPP_MASS'],10**SD['COUPP_XSEC'],
                        label='COUPP',linewidth='2',color="purple")
icecube_bb = plt.loglog(10**SD['ICECUBEbb_MASS'],10**SD['ICECUBEbb_XSEC'],
                        label='ICECUBE (bb)',linewidth='2',linestyle='--',color='midnightblue')
icecube_WW = plt.loglog(10**SD['ICECUBEWW_MASS'],10**SD['ICECUBEWW_XSEC'],
                        label='ICECUBE (W$^+$W$^-$)',linewidth='2',linestyle=':',color='midnightblue')

lines = [cdms, xenon10, xenon100, picasso, simple, coupp, icecube_bb, icecube_WW]

### Add legend
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.savefig('SD.png', bbox_inches='tight')
```

#Spin independent

```python
SI = pd.read_csv("SI.csv",sep=",",header=None,names=range(100)).T
SI.columns = [SI[i][0] for i in SI]
SI = SI.ix[1:]
```

```python
SI.columns.values.tolist()[0::2]
```

```python
import math

def not_nan(ar):
    """Takes a series and returns all elements that are not NaN."""
    return np.array([el for el in np.asarray(ar) if not math.isnan(el)])
```

```python
### Create figure and basic styling
fig = plt.figure(dpi=1200)
ax = fig.add_subplot(111,xlabel='WIMP mass [GeV]',ylabel='WIMP-nucleon cross-section [cm$^2$]')

for item in [ax.xaxis.label, ax.yaxis.label]:
    item.set_fontsize(16)

for item in (ax.get_xticklabels() + ax.get_yticklabels()):
    item.set_fontsize(14)

plt.xlim([3,10**3])
plt.ylim([10**-46,10**-39])

    
### Create exclusion lines
lux =        plt.loglog(10**np.append(not_nan(SI['LUXLOWMASS_MASS']),not_nan(SI['LUXHIGHMASS_MASS'][2:])),
                        10**np.append(not_nan(SI['LUXLOWMASS_XSEC']),not_nan(SI['LUXHIGHMASS_XSEC'][2:])),
                        label='LUX',linewidth='2',color="midnightblue")
xenon100 =   plt.loglog(10**SI['XENON100_MASS'],10**SI['XENON100_XSEC'],
                        label='XENON100',linewidth='2',color="tomato")
xenon10 =    plt.loglog(10**SI['XENON10_MASS'],10**SI['XENON10_XSEC'],
                        label='XENON10',linewidth='2',color="tomato",linestyle="--")
cdmsGe =     plt.loglog(10**SI['CDMSGe_MASS'],10**SI['CDMSGe_XSEC'],
                        label='CDMS II (Ge)',linewidth='2',color="olivedrab")
simple =     plt.loglog(10**SI['SIMPLE_MASS'],10**SI['SIMPLE_XSEC'],
                        label='SIMPLE',linewidth='2',color="steelblue",linestyle="--")
couppLower = plt.loglog(10**SI['COUPPlower_MASS'],10**SI['COUPPlower_XSEC'],
                        label='COUPP',linewidth='2',color="purple")

### The lines for the hints (and upper 90% CL on COUPP)

#couppUpper = plt.loglog(10**SI['COUPPupper_MASS'],10**SI['COUPPupper_XSEC'],label='COUPP',linewidth='2') # lower and upper represent 90% CL
#dama1 = plt.loglog(10**SI['DAMA1_MASS'],10**SI['DAMA1_XSEC'],label='DAMA',color='#1d706a',linewidth='1')
#dama2 = plt.loglog(10**SI['DAMA2_MASS'],10**SI['DAMA2_XSEC'],color='#1d706a',linewidth='1')
#cresst1 = plt.loglog(10**SI['CRESST1_MASS'],10**SI['CRESST1_XSEC'],label='CRESST',color='#10403c',linewidth='1')
#cresst2 = plt.loglog(10**SI['CRESST2_MASS'],10**SI['CRESST2_XSEC'],color='#10403c',linewidth='1')
#cdmsSi, = plt.loglog(10**SI['CDMSSi_MASS'],10**SI['CDMSSi_XSEC'],label='CDMS (Si)',color='darkblue',linewidth='1')

### Create patches for "hint" areas
dama1_patch = np.array([not_nan(10**SI['DAMA1_MASS']),not_nan(10**SI['DAMA1_XSEC'])])
dama1_patch = dama1_patch.T
dama1_patch = pat.Polygon(dama1_patch,color='cornflowerblue',alpha=0.4,closed=True,label = 'DAMA')

dama2_patch = np.array([not_nan(10**SI['DAMA2_MASS']),not_nan(10**SI['DAMA2_XSEC'])])
dama2_patch = dama2_patch.T
dama2_patch = pat.Polygon(dama2_patch,color='cornflowerblue',alpha=0.4)

cresst1_patch = np.array([not_nan(10**SI['CRESST1_MASS']),not_nan(10**SI['CRESST1_XSEC'])])
cresst1_patch = cresst1_patch.T
cresst1_patch = pat.Polygon(cresst1_patch,color='indianred',alpha=0.4,label = 'CRESST')

cresst2_patch = np.array([not_nan(10**SI['CRESST2_MASS']),not_nan(10**SI['CRESST2_XSEC'])])
cresst2_patch = cresst2_patch.T
cresst2_patch = pat.Polygon(cresst2_patch,color='indianred',alpha=0.4)

cdmsSi_patch = np.array([not_nan(10**SI['CDMSSi_MASS']),not_nan(10**SI['CDMSSi_XSEC'])])
cdmsSi_patch = cdmsSi_patch.T
cdmsSi_patch = pat.Polygon(cdmsSi_patch,color='olivedrab',alpha=0.4,label = 'CDMS (Si)')

lines = [lux,xenon100, xenon10, cdmsGe, simple,couppLower]
patches = [dama1_patch,dama2_patch,cresst1_patch,cresst2_patch,cdmsSi_patch]

### Add patches to plot
p = PatchCollection(patches,match_original=True)
ax.add_collection(p)

### Label the patches
ax.text(0.17, 0.89, 'DAMA', 
        verticalalignment='bottom', horizontalalignment='left',
        transform=ax.transAxes, color='cornflowerblue', fontsize=12)
ax.text(0.46, 0.68, 'DAMA', 
        verticalalignment='bottom', horizontalalignment='left',
        transform=ax.transAxes, color='cornflowerblue', fontsize=12)
ax.text(0.39, 0.57, 'CRESST', 
        verticalalignment='center', horizontalalignment='center',
        transform=ax.transAxes, color='indianred', fontsize=12, rotation=-40)
ax.text(0.23, 0.795, 'CRESST', 
        verticalalignment='center', horizontalalignment='center',
        transform=ax.transAxes, color='indianred', fontsize=8, rotation=-49)
ax.annotate('CDMS II (Si)', xy=(0.26, 0.71), xytext=(0.30, 0.76),
            arrowprops=dict(arrowstyle="->",color='olivedrab',relpos=(0.0,0.1),patchA=None),
            xycoords='axes fraction',color='olivedrab',
            fontsize=12, verticalalignment='bottom ', horizontalalignment='left')

### Add legend for the lines
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

### Output in vector and raster form
plt.savefig('SI.png', bbox_inches='tight')
plt.savefig('SI.pdf', bbox_inches='tight')
```
